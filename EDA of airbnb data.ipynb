```python
import pandas as pd
```

```python
#load airbnb data
airbnb_data = pd.read_csv("C:\\Users\\user\\Desktop\\airbnb\\listings.csv")
```

```python
airbnb_data.head()
```

```python
# Basic data exploration
print(airbnb_data.head()) 
```

```python
# Get data information
print(airbnb_data.info())  
```

```python
# Summary statistic
print(airbnb_data.describe())
```

# Descriptive Analysis

```python
average_price = airbnb_data['price'].mean()
print(f'Average Price: ${average_price:.2f}')
```

```python
most_common_room_type = airbnb_data['room_type'].mode().iloc[0]
print(f'Most Common room type: & {most_common_room_type}')
```

```python
#calculate the average reviews per month
average_reviews_per_month = airbnb_data['number_of_reviews'].mean()
print(f'average reveiws per month: &{average_reviews_per_month}')
```

# Geographical Analysis

```python
top_neighborhoods = airbnb_data['neighbourhood'].value_counts().head(10)
print('Top Neighborhoods by Listing Count:')
print(top_neighborhoods)
```

# Time Series analysis

```python
# Convert 'last_review' to datetime
airbnb_data['last_review'] = pd.to_datetime(airbnb_data['last_review'])
```

```python
# Group by year and month
availability_over_time = airbnb_data.groupby(airbnb_data['last_review'].dt.to_period("M"))['availability_365'].mean()
print('Availability Over Time:')
print(availability_over_time)
```

```python
#explore seasonal price trends
price_over_time = airbnb_data.groupby(airbnb_data['last_review'].dt.to_period("M"))['price'].mean().head(10)
print('Price Over Time:')
print(price_over_time)
```

# Host analysis

```python
top_hosts = airbnb_data['host_name'].value_counts().head(5)
print('Top Hosts by Listing Count:')
print(top_hosts)
```

```python
#average listing per host
average_listings_per_host = airbnb_data['calculated_host_listings_count'].mean()
print(f'Average Listings Per Host: {average_listings_per_host:.2f}')
```

# Descriptive analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#price distribution histogram
plt.figure(figsize=(10, 6))
sns.histplot(airbnb_data['price'], bins=50, kde=True)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Price Distribution')
plt.show()
```

```python
#room type distribution
plt.figure(figsize=(10, 6))
sns.histplot(airbnb_data['room_type'], bins=50, kde=True)
plt.xlabel('room type')
plt.ylabel('count')
plt.title('Room type Distribution')
plt.show()
```

```python
#Reviews per Month Distribution
plt.figure(figsize=(10, 6))
sns.histplot(airbnb_data['reviews_per_month'].dropna(), bins=50, kde=True)
plt.xlabel('Reviews Per Month')
plt.ylabel('Frequency')
plt.title('Reviews Per Month Distribution')
plt.show()
```

# Geographical analysis

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_neighborhoods.index, y=top_neighborhoods.values)
plt.xlabel('Neighborhood')
plt.ylabel('Listing Count')
plt.title('Top Neighborhoods by Listing Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#timeseries visuals
plt.figure(figsize=(12, 6))
availability_over_time.plot(kind='line', marker='o')
plt.xlabel('Time (Year-Month)')
plt.ylabel('Average Availability (365 days)')
plt.title('Availability Over Time')
plt.grid(True)
plt.show()
```

```python
#price over time
plt.figure(figsize=(12, 6))
price_over_time.plot(kind='line', marker='o', color='purple')
plt.xlabel('Time (Year-Month)')
plt.ylabel('Average Price')
plt.title('Price Over Time')
plt.grid(True)
plt.show()
```

```python
#host analysis
plt.figure(figsize=(10, 6))
sns.barplot(x=top_hosts.index, y=top_hosts.values)
plt.xlabel('Host Name')
plt.ylabel('Listing Count')
plt.title('Top Hosts by Listing Count')
plt.xticks(rotation=45)
plt.show()
```
